```python
# 1. kütüphaneler
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# 2. web driver yüklenmesi (chrome 106 için)
driver = webdriver.Chrome(executable_path="C:/Users/Vedat/anaconda3/chromedriver.exe")
```

```python
# 3. Arama bilgisi
sahibinden_arama = "Hyundai Getz"
```

```python
# 4. site adresine gidelim
driver.get("https://www.sahibinden.com")
sleep(5)
```

```python
# 5. cookie kapat (eğer varsa)
driver.find_element_by_xpath('//*[@id="onetrust-accept-btn-handler"]')\
    .click()
sleep(5)
```

```python
# 6. arama kutusu
driver.find_element_by_xpath('//*[@id="searchText"]')\
    .send_keys(sahibinden_arama)
sleep(5)
```

```python
# 7. arama butonuna bas
driver.find_element_by_xpath('//*[@id="searchSuggestionForm"]/button')\
    .click()
sleep(5)
```

```python
# 8. Arama sayısını 50 yap
# Burada <LI> etiketi kullanıldığı için bu şekilde butona ulaşıyoruz.
driver.execute_script("arguments[0].click();", WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="searchResultsSearchForm"]/div/div[3]/div[3]/div[2]/ul/li[2]/a'))))
sleep(5)
```

```python
# 9. Çekilecek bilgilerin değişkenleri
item_titles = driver.find_elements_by_xpath('//*[@id="searchResultsTable"]/tbody/tr[*]/td[2]/a')
item_prices = driver.find_elements_by_xpath('//*[@id="searchResultsTable"]/tbody/tr[*]/td[3]/div/span')
item_loca = driver.find_elements_by_xpath('//*[@id="searchResultsTable"]/tbody/tr[*]/td[5]')
item_date = driver.find_elements_by_xpath('//*[@id="searchResultsTable"]/tbody/tr[*]/td[4]')
```

```python
# 10. boş listler
titles_list = []
prices_list = []
loca_list = []
date_list = []
```

```python
# 11.for döngüsü ile ürünlerin istediğimiz bilgileri alma

for title in item_titles:
    titles_list.append(title.text)

for price in item_prices:
    prices_list.append(price.text)

for locat in item_loca:
    loca_list.append(locat.text)

for dat in item_date:
    date_list.append(dat.text)

# ekranı kapatalım
driver.close()
```

```python
# 12. listelere bakalım
print(titles_list)
print(prices_list)
print(loca_list)
print(date_list)
```

```python
# 13. Dataframe oluşturalım.
dfS = pd.DataFrame(zip(titles_list, prices_list, loca_list, date_list), columns=["İlan Başlığı", "Fiyat", "Lokasyon", "İlan Tarihi"])
dfS["Fiyat"] = dfS["Fiyat"].str.replace(",", ".")
dfS["Fiyat"] = dfS["Fiyat"].str.replace(" TL", "").astype(float)
dfS
```

```python
# Ortalama fiyat
print(dfS["Fiyat"].mean())
```

```python
# Minfiyat
print(dfS["Fiyat"].min())
```

```python
# Max Fiyat
print(dfS["Fiyat"].max())
```

```python
# detay
print(dfS["Fiyat"].describe())
```
