```python
class Dosya(): 

    def __init__(self):
        # dosyayı okuyalım
        with open ("metin.txt", "r", encoding="utf-8") as file:
            dosya_icerigi = file.read()
            kelimeler = dosya_icerigi.split() # dosya parçalayıp içeriğini lite haline getirdik
            self.sade_kelimeler = list()
            for i in kelimeler:
                i = i.strip("\n") # \n atalım
                i = i.strip(" ") # boşluk atalım
                i = i.strip(".") # . azaltalım
                i = i.strip(",") # , atalım
                i = i.strip('"') # " atalım
                self.sade_kelimeler.append(i)
    
    def tum_kelimeler(self): 
        kelimeler_kumesi = set()
        for i in self.sade_kelimeler:
            kelimeler_kumesi.add(i) # tüm kelimelerden küme oluşturduk. Her kelime tek oldu
        print("Tüm Kelimeler .......")
        for i in kelimeler_kumesi:
            print(i)
            print("*" * 25)

    def kelime_frekansi(self): 
        kelime_sozluk = dict()
        for i in self.sade_kelimeler:
            if (i in kelime_sozluk):
                kelime_sozluk[i] += 1
            else:
                kelime_sozluk[i] = 1
        for kelime, sayi in kelime_sozluk.items():
            print("{} - kelimesi {} defa geçiyor ... ".format(kelime, sayi))
            print("-" * 45)

            
```

```python
dosya = Dosya()
dosya.kelime_frekansi() 
```
