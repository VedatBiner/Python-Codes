```python
# Her bir videodaki etiket sayısını hesaplayarak, 
# oluşan değerleri yeni sütun olarak ekleyelim. 
def countTag(tag):
    tagList = tag.split("|")
    return len(tagList)
```

```python
youtube["tag_count"] = youtube["tags"].apply(countTag)
```

```python
# Videoları, beğeni oranı en yüksekten en düşüğe doğru sıralayalım.
# iteritems() ile tupple olarak alıp, listeye çevirelim,
# Index değerlerini de atalım. Liste de for döngüsü ile gezinelim.
# likes ve dislikes değerlerini aynı anda alan bir fonksiyon oluşturalım.
# bu fonksiyon ile elde ettiğimiz oranları yeni sütuna ekleyelim.
# ve bu sütuna göre büyükten küçüğe sıralayıp yeni tablo oluşturalım
def likesdislikesratio(likes, dislikes):
    likesList = []
    dislikesList = []
    ratio = []
    for key, value in list(likes.iteritems()):
        likesList.append(value)
    for key, value in list(dislikes.iteritems()):
        dislikesList.append(value)
    likedislike = list(zip(likesList, dislikesList))
    for like, dislike in likedislike:
        if (like + dislike) == 0:
            ratio.append(0)
        else:
            ratio.append(like / (like + dislike))
    return ratio
```

```python
youtube["likes_dislikes"] = likesdislikesratio(youtube["likes"], youtube["dislikes"])
```

```python
youtube.sort_values("likes_dislikes", ascending = False, inplace = True)
```
